```python
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

def create_pdf_report(file_name, left_data, right_data):
    doc = SimpleDocTemplate(file_name, pagesize=letter)
    elements = []

    table_left = Table(left_data)
    table_left.setStyle(TableStyle([('GRID', (0, 0), (-1, -1), 0, colors.white)]))

    table_right = Table(right_data)
    table_right.setStyle(TableStyle([('GRID', (0, 0), (-1, -1), 0, colors.white)]))

    elements.append(table_left)
    elements.append(table_right)

    styles = getSampleStyleSheet()

    section1_text = f"Report Date: {reportdate}"
    section1 = Paragraph(section1_text, styles["Normal"])

    section1_righttext = f"<para align=right>Source room: {source_room_name}</para>"
    section1_1 = Paragraph(section1_righttext, styles["Normal"])

    section2_text = f"Test Date: {testdate}<br/>Test Site: {sitename}<br/>Client Name: {client_Name}"
    section2 = Paragraph(section2_text, styles["Normal"])

    elements.extend([section1, section1_1, section2])

    # Build the PDF
    doc.build(elements)

# Define your variables here
reportdate = "4-24-24"
testdate = "4-3-24"
source_room_name = "2nd Floor Bed 3"
sitename = "Ka'ulu by Gentry"
client_Name = "Gentry Builders, LLC"

leftside_data = [
    ["Report Date:", reportdate],
    ['Test Date:', testdate]
]

rightside_data = [
    ["Source Room:", source_room_name],
    ["Test Site:", sitename],
    ["Client Name:", client_Name]
]

# Generate the PDF report
create_pdf_report("example_report.pdf", leftside_data, rightside_data)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import BaseDocTemplate, PageTemplate, Frame, Paragraph, Spacer, PageBreak
from reportlab.pdfgen.canvas import Canvas

# Custom margins and frame dimensions
left_margin = right_margin = 0.75 * 72  # 0.75 inches
top_margin = 0.25 * 72  # 0.25 inch
bottom_margin = 1 * 72  # 1 inch

header_height = 1 * inch
footer_height = 0.5 * inch

# Initialize BaseDocTemplate
doc = BaseDocTemplate("report_with_header_footer.pdf", pagesize=letter,
                      leftMargin=left_margin, rightMargin=right_margin,
                      topMargin=top_margin, bottomMargin=bottom_margin)

# Define Frames for the header, main content, and footer
header_frame = Frame(left_margin, letter[1] - top_margin - header_height, letter[0] - 2 * left_margin, header_height, id='header')
main_frame = Frame(left_margin, bottom_margin + footer_height, letter[0] - 2 * left_margin, letter[1] - top_margin - bottom_margin - header_height - footer_height, id='main')
footer_frame = Frame(left_margin, bottom_margin, letter[0] - 2 * left_margin, footer_height, id='footer')

# Define a function to draw header and footer
def header_footer(canvas, doc):
    canvas.saveState()
    canvas.setFont('Helvetica-Bold', 14)
    
    # Multiple lines of centered text for the header
    header_text = ["Field Impact Sound Transmission Test Report",
                   "Apparent Impact Insulation Class (AIIC)"]
    header_y_positions = [letter[1] - top_margin - header_height / 2 + 10, letter[1] - top_margin - header_height / 2 - 10]

    for text, y in zip(header_text, header_y_positions):
        canvas.drawCentredString(letter[0] / 2, y, text)

    # Footer
    canvas.setFont('Helvetica', 10)
    footer_text = f"Page {doc.page}"
    canvas.drawCentredString(letter[0] / 2, bottom_margin + footer_height / 2, footer_text)

    canvas.restoreState()


# Create a page template with header and footer
page_template = PageTemplate(id='Standard', frames=[header_frame, main_frame, footer_frame], onPage=header_footer)
doc.addPageTemplates([page_template])

# Styles and main document content
styles = getSampleStyleSheet()
elements = [
    Paragraph("This is the main body text on the first page.", styles['Normal']),
    PageBreak(),
    Paragraph("This is text on the second page after a page break.", styles['Normal'])
]

# Build the document
doc.build(elements)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import BaseDocTemplate, PageTemplate, Frame, Paragraph, Spacer, PageBreak
from reportlab.pdfgen.canvas import Canvas

# Custom margins and frame dimensions
left_margin = right_margin = 0.75 * 72  # 0.75 inches
top_margin = 0.25 * 72  # 0.25 inch
bottom_margin = 1 * 72  # 1 inch

header_height = 1 * inch
footer_height = 0.5 * inch


main_content_height = letter[1] - top_margin - bottom_margin - header_height - footer_height

# Initialize BaseDocTemplate
doc = BaseDocTemplate("report_with_header_footer.pdf", pagesize=letter,
                      leftMargin=left_margin, rightMargin=right_margin,
                      topMargin=top_margin, bottomMargin=bottom_margin)

# Define Frames for the header, main content, and footer
header_frame = Frame(left_margin, letter[1] - top_margin - header_height, letter[0] - 2 * left_margin, header_height, id='header')
main_frame = Frame(left_margin, bottom_margin + footer_height, letter[0] - 2 * left_margin, main_content_height, id='main')
footer_frame = Frame(left_margin, bottom_margin, letter[0] - 2 * left_margin, footer_height, id='footer')

# Define a function to draw header and footer
def header_footer(canvas, doc):
    canvas.saveState()
    canvas.setFont('Helvetica', 10)
    canvas.drawCentredString(letter[0] / 2, letter[1] - top_margin - (header_height / 2) + 15, "Field Impact Sound Transmission Test Report")
    canvas.drawCentredString(letter[0] / 2, bottom_margin - (footer_height / 2) + 10, f"Page {doc.page} of 4")
    canvas.restoreState()

# Create a page template with header and footer
page_template = PageTemplate(id='Standard', frames=[header_frame, main_frame, footer_frame], onPage=header_footer)
doc.addPageTemplates([page_template])

# Styles and main document content
styles = getSampleStyleSheet()
elements = [
    Paragraph("This is the main body text on the first page.", styles['Normal']),
    PageBreak(),
    Paragraph("This is text on the second page after a page break.", styles['Normal'])
]

# Build the document
doc.build(elements)
```
