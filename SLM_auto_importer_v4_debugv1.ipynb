```python
# Create a PDF document
source_room_name = "2nd Floor Bed 3"
sitename = "Ka'ulu by Gentry"
client_Name = "Gentry Builders, LLC"

def create_pdf_report(file_name):
# Create a PDF document
    doc = SimpleDocTemplate(file_name, pagesize=letter)
    elements = []

    # Add sections of text with Arial font
    section1_text = f"Report Date: {reportdate}"
    section1 = Paragraph(section1_text, getSampleStyleSheet()["Normal"])
    elements.append(section1)

    section1_righttext = f"<para align=right>Source room: {source_room_name}</para>"
    section1_1 = Paragraph(section1_righttext, getSampleStyleSheet()["Normal"])
    elements.append(section1_1)
    
    section2_text = f"Test Date:{testdate}<br/>Test Site: {sitename}\nClient Name:{client_Name}"
    section2 = Paragraph(section2_text, getSampleStyleSheet()["Normal"])
    elements.append(section2)

    # Add elements to the document
    doc.build(elements)


# Generate the PDF report
create_pdf_report("example_report.pdf")
```
